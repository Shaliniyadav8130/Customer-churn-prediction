```python
df1 = df.drop(['customerID'], axis=1)
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols = ['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','Churn']
df1[cols] = df1[cols].apply(LabelEncoder().fit_transform)
df1.head()
```

```python
df1['MonthlyCharges'] = df1['MonthlyCharges'].astype(int)
```

```python
l1 = [len(i.split()) for i in df1['TotalCharges']]
l2 = [i for i in range(len(l1)) if l1[i] != 1]
print('Index Positions with empty spaces : ',*l2)

for i in l2:
    df1.loc[i,'TotalCharges'] = df1.loc[(i-1),'TotalCharges']
    
df1['TotalCharges'] = df1['TotalCharges'].astype(float)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```
